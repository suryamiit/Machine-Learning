```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn import linear_model  # importing to do linear regression
%matplotlib inline 
# visualize the graph in result itself we included the line
```

```python
# read data set
dataset_train = pd.read_csv('father_son_heights.csv')


# assign data into panda data frames
x_train = dataset_train['Father'].values.reshape(-1,1)
#reshaping the 1D array to 2 D to make it easy for sklearn. List to array

y_train = dataset_train['Son'].values.reshape(-1,1)
#print ("Training data \n",dataset_train)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.savefig('scatterplot.png')
```

```python
# calling linear regression function
reg = linear_model.LinearRegression()
res= reg.fit(x_train, y_train)
```

```python
print('Predicted Coefficient Values =', reg.coef_)
print('Predicted Intercept Value= ',reg.intercept_)
f=open("Linearreg.txt","x")
f=open("Linearreg.txt","a")
f.write("Predicted Coefficient Value = ")
f.write("%s\n" % reg.coef_)
f.write("Predicted Intercept Value = ")
f.write("%s\n" % reg.intercept_)
f.close()
```

```python
# Linspace gives equally spaced points between the 2 parameters
x_viz = np.linspace(x_train.min(),x_train.max())
plt.scatter(x_train,y_train)
#plt.plot(x_viz,reg.intercept_+reg.coef_*x_viz,color='r') 
plt.plot(x_viz,reg.predict(x_viz.reshape(-1,1)), color='r')
# y = wo+w1*x 
plt.savefig('line_and_scatter.png')
plt.show()
#plt.savefig('line_and_scatter.png')
```
